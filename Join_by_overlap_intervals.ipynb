```python
import pandas as pd
```

```python
ceps = pd.read_excel("cepcid.xlsx")
cobertura = pd.read_excel("cepcid.xlsx",sheet_name="Planilha2")
# ceps["dummy"] = 1
# cobertura["dummy"]= 1
```

```python
ceps["intervalo"] = [pd.Interval(row["CEPMIN"], row["CEPMAX"], "both") for index, row in ceps.iterrows()]
cobertura["intervalo"] = [pd.Interval(row["CEPMIN"], row["CEPMAX"], "both") for index, row in cobertura.iterrows()]
```

```python
ceps.head()
```

```python
cobertura.head()
```

```python
df_cross = ceps.merge(cobertura, how="cross", suffixes=['_ceps','_cobertura'])
df_cross.head()
```

```python
df_cross["test"] = [row["intervalo_ceps"].overlaps(row["intervalo_cobertura"]) for index, row in df_cross.iterrows()]
```

```python
df_cross.head()
```

```python
(
    df_cross
    .query("test==True")
    .loc[:,["CIDADE_OFICIAL", "CEPMIN_cobertura", "CEPMAX_cobertura","CLUSTER"]]
)
```
